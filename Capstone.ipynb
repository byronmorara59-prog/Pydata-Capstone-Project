```python
import yfinance as yf
import pandas as pd
```

```python
df_brent = yf.download("BZ=F", start="2007-07-30", end="2025-12-31")
```

```python
df_gas   = yf.download("NG=F", start="2007-07-30", end="2025-12-31")
```

```python
df_dxy   = yf.download("DX-Y.NYB", start="2007-07-30", end="2025-12-31")
```

```python
df_dxy.info()
```

```python
df_merged = pd.concat([df_brent, df_gas, df_dxy], axis=1, join="inner")
```

```python
df_merged.shape
```

```python
df_merged.head()
```

```python
#Flattening the MultiIndex columns
df_merged.columns = ['_'.join(col) for col in df_merged.columns]
```

```python
#Confirm
df_merged.columns.tolist()
```

```python
df_merged.rename(columns={
    #Brent
    'Close_BZ=F'  : 'Brent_Close',
    'High_BZ=F'   : 'Brent_High',
    'Low_BZ=F'    : 'Brent_Low',
    'Open_BZ=F'   : 'Brent_Open',
    'Volume_BZ=F' : 'Brent_Volume',
    #Natural Gas
    'Close_NG=F'  : 'Gas_Close',
    'High_NG=F'   : 'Gas_High',
    'Low_NG=F'    : 'Gas_Low',
    'Open_NG=F'   : 'Gas_Open',
    'Volume_NG=F' : 'Gas_Volume',
    #S&P 500
    'Close_^GSPC' : 'SP500_Close',
    'High_^GSPC'  : 'SP500_High',
    'Low_^GSPC'   : 'SP500_Low',
    'Open_^GSPC'  : 'SP500_Open',
    'Volume_^GSPC': 'SP500_Volume',
    #US Dollar Index
    'Close_DX-Y.NYB' : 'DXY_Close',
    'High_DX-Y.NYB'  : 'DXY_High',
    'Low_DX-Y.NYB'   : 'DXY_Low',
    'Open_DX-Y.NYB'  : 'DXY_Open',
    'Volume_DX-Y.NYB': 'DXY_Volume'
}, inplace=True)
```

```python
df_merged.columns.tolist()
```

```python
df_merged.head()
```

```python
df_merged['Brent_Return']     = df_merged['Brent_Close'].pct_change() * 100
df_merged['Brent_Range']      = df_merged['Brent_High'] - df_merged['Brent_Low']
df_merged['Brent_MA7']        = df_merged['Brent_Close'].rolling(window=7).mean()
df_merged['Brent_MA30']       = df_merged['Brent_Close'].rolling(window=30).mean()
df_merged['Brent_Volatility'] = df_merged['Brent_Return'].rolling(window=30).std()
df_merged.head(35)
```

```python
import matplotlib.pyplot as plt
```

```python
df_monthly = df_merged['Brent_Close'].resample('ME').mean()
```

```python
#df_monthly = df_merged['Brent_Close'].resample('ME').mean()
ma12 = df_monthly.rolling(window=12).mean()

fig, ax = plt.subplots(figsize=(14, 5))

ax.plot(df_monthly.index, df_monthly, 
        color='steelblue', linewidth=1.2, 
        label='Monthly Avg Price')

ax.plot(ma12.index, ma12, 
        color='red', linewidth=2, 
        label='12-Month MA')

ax.set_title('Brent Crude Oil Monthly Price Trends (2007 - 2025)')
ax.set_xlabel('Date')
ax.set_ylabel('Price (USD per Barrel)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show();
```

```python
fig, ax = plt.subplots(figsize=(14, 5))

ax.plot(df_merged.index, df_merged['Brent_Volatility'],
        color='red', linewidth=1.5)

ax.set_title('Brent Crude Oil Monthly Rolling Volatility (2007 - 2025)')
ax.set_xlabel('Date')
ax.set_ylabel('Volatility')
ax.grid(True, alpha=0.3)
plt.show();
```

```python
import seaborn as sns

df_corr = df_merged[['Brent_Close', 'Gas_Close', 'DXY_Close']]

corr_matrix = df_corr.corr()

fig, ax = plt.subplots(figsize=(8, 6))

sns.heatmap(corr_matrix, 
            annot=True, 
            fmt='.2f',
            cmap='coolwarm',
            ax=ax)

ax.set_title('Correlation Matrix — Brent Crude and Market Indicators')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

#Brent against Gas
axes[0].scatter(df_merged['Gas_Close'], df_merged['Brent_Close'],
                alpha=0.3, color='steelblue', s=5)
axes[0].set_title('Brent vs Natural Gas')
axes[0].set_xlabel('Natural Gas Price')
axes[0].set_ylabel('Brent Price (USD)')

#Brent against DXY
axes[1].scatter(df_merged['DXY_Close'], df_merged['Brent_Close'],
                alpha=0.3, color='red', s=5)
axes[1].set_title('Brent vs US Dollar Index')
axes[1].set_xlabel('DXY Index')
axes[1].set_ylabel('Brent Price (USD)')

plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

df_monthly = df_merged['Brent_Close'].resample('ME').mean()
ma12 = df_monthly.rolling(window=12).mean()

fig = go.Figure()

#Monthly Price line
fig.add_trace(go.Scatter(
    x=df_monthly.index,
    y=df_monthly,
    name='Monthly Avg Price',
    line=dict(color='steelblue', width=1.5)
))

#12 month Moving Average
fig.add_trace(go.Scatter(
    x=df_monthly.index,
    y=ma12,
    name='12-Month MA',
    line=dict(color='red', width=2)
))

#Event markers
events = {
    '2008 Financial Crisis' : '2008-09-15',
    '2015-16 Supply Glut'   : '2015-12-01',
    'COVID-19 Crash'        : '2020-03-18',
    'Russia-Ukraine War'    : '2022-02-24'
}

for label, date in events.items():
    fig.add_vline(
        x=date,
        line_dash='dash',
        line_color='black',
        line_width=1,
        opacity=0.7
    )
    fig.add_annotation(
        x=date,
        y=df_monthly.max() * 0.98,
        text=label,
        showarrow=False,
        textangle=-90,
        font=dict(size=10, color='black'),
        bgcolor='yellow',
        opacity=0.7
    )

fig.update_layout(
    title='Brent Crude Oil Monthly Price Trends (2007 - 2025)',
    xaxis=dict(
        title='Date',
        rangeslider=dict(visible=True),
        type='date'
    ),
    yaxis=dict(title='Price (USD per Barrel)'),
    legend=dict(x=0, y=1.1, orientation='h'),
    height=600
)

fig.show()
```
